```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random


# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"


# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
merge_tab = pd.merge(ride_data, city_data, on="city", how="left")

# Display the data table for preview
```

```python
percentages = 100*merge_tab.groupby(["type"]).sum()["fare"]/merge_tab["fare"].sum()
percentages
```

## Total Fares by City Type

```python
# Calculate Type Percents
labels =["Rural", "Urban","Suburban"]
explode =(0,0,0.1)
colors = ["lightcoral","lightskyblue","orange"]
plt.title("% of Total Fares by City Type")
plt.pie(percentages, labels=labels,colors=colors,autopct = "%1.2f%%", explode=explode,shadow=True,startangle=140)

# Build Pie Chart

# Save Figure
#plt.savefig("")
```

## Total Rides by City Type

```python
# Calculate Ride Percents
percentages1 = 100*merge_tab.groupby(["type"]).sum()["ride_id"]/merge_tab["ride_id"].sum()
# Build Pie Chart
labels =["Rural", "Urban","Suburban"]
explode =(0,0,0.1)
colors = ["lightcoral","lightskyblue","orange"]
plt.title("% of Total Rides by City Type")
plt.pie(percentages1, labels=labels,colors=colors,autopct = "%1.2f%%", explode=explode,shadow=True,startangle=140)

# Save Figure
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
percentages2 = 100*merge_tab.groupby(["type"]).sum()["driver_count"]/merge_tab["driver_count"].sum()
# Build Pie Charts
labels =["Rural", "Urban","Suburban"]
explode =(0,0,0.1)
colors = ["lightcoral","lightskyblue","orange"]
plt.title("% of Total Drivers by City Type")
plt.pie(percentages2, labels=labels,colors=colors,autopct = "%1.2f%%", explode=explode,shadow=True,startangle=140)
# Save Figure
```
